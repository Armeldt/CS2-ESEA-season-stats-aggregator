```python
import pandas as pd
from demoparser2 import DemoParser

# Chemin vers la démo
path_to_demo = "demos/VS_Eldsjalar_Vertigo.dem"

def find_if_1vx(deaths, round_idx, round_ends, df, X):
    for _, death in deaths.iterrows():
        if death["total_rounds_played"] == round_idx:
            subdf = df[df["tick"] == death["tick"]]
            ct_alive = subdf[(subdf["team_name"] == "CT") & (subdf["is_alive"] == True)]
            t_alive = subdf[(subdf["team_name"] == "TERRORIST") & (subdf["is_alive"] == True)]
            # 3 = CT
            if len(ct_alive) == 1 and len(t_alive) == X and round_ends.iloc[round_idx]["winner"] == 3:
                return ct_alive["name"].iloc[0]
            # 2 = T
            if len(t_alive) == 1 and len(ct_alive) == X and round_ends.iloc[round_idx]["winner"] == 2:
                return t_alive["name"].iloc[0]

# Initialiser un dictionnaire pour stocker le nombre de clutchs 1vX par joueur
clutches_1vx = {}

parser = DemoParser(path_to_demo)
deaths = parser.parse_event("player_death", other=["total_rounds_played"])
round_ends = parser.parse_event("round_end")
df2 = parser.parse_ticks(["is_alive", "team_name", "team_rounds_total"], ticks=deaths["tick"].to_list())
max_round = deaths["total_rounds_played"].max() + 1

# Parcourir chaque round dans la démo
for X in range(1, 6):
    for round_idx in range(0, max_round):
        clutcher_steamid = find_if_1vx(deaths, round_idx, round_ends, df2, X)
        if clutcher_steamid is not None:
            # Mise à jour du nombre de clutchs 1vX par joueur dans le dictionnaire
            clutches_1vx.setdefault(clutcher_steamid, {}).setdefault(f"1v{X}", 0)
            clutches_1vx[clutcher_steamid][f"1v{X}"] += 1

# Convertir le dictionnaire en DataFrame pandas
clutches_df = pd.DataFrame.from_dict(clutches_1vx, orient="index").fillna(0)
clutches_df.index.name = "Player"
print(clutches_df)
```

```python
import os
import pandas as pd
from demoparser2 import DemoParser

# Fonction pour trouver le joueur qui a réussi un clutch 1vX
def find_if_1vx(deaths, round_idx, round_ends, df, X, clutches_1vx):
    for _, death in deaths.iterrows():
        if death["total_rounds_played"] == round_idx:
            subdf = df[df["tick"] == death["tick"]]
            ct_alive = subdf[(subdf["team_name"] == "CT") & (subdf["is_alive"] == True)]
            t_alive = subdf[(subdf["team_name"] == "TERRORIST") & (subdf["is_alive"] == True)]
            # 3 = CT
            if len(ct_alive) == 1 and len(t_alive) == X and round_ends.iloc[round_idx]["winner"] == 3:
                return ct_alive["name"].iloc[0]
            # 2 = T
            if len(t_alive) == 1 and len(ct_alive) == X and round_ends.iloc[round_idx]["winner"] == 2:
                return t_alive["name"].iloc[0]

# Dossier contenant les démos
demo_folder = "demos"

# Initialiser une liste de dictionnaires pour stocker le nombre de clutchs 1vX par joueur
clutches_1vx_list = []

# Parcourir chaque fichier dans le dossier de démos
for filename in os.listdir(demo_folder):
    if filename.endswith(".dem"):
        path_to_demo = os.path.join(demo_folder, filename)
        print(f"Analyzing demo: {filename}")
        
        parser = DemoParser(path_to_demo)
        deaths = parser.parse_event("player_death", other=["total_rounds_played"])
        round_ends = parser.parse_event("round_end")
        df2 = parser.parse_ticks(["is_alive", "team_name", "team_rounds_total"], ticks=deaths["tick"].to_list())
        max_round = deaths["total_rounds_played"].max() + 1

        # Parcourir chaque round dans la démo
        for X in range(1, 6):
            for round_idx in range(0, max_round):
                clutcher_steamid = find_if_1vx(deaths, round_idx, round_ends, df2, X, clutches_1vx_list)
                if clutcher_steamid is not None:
                    # Mise à jour du nombre de clutchs 1vX par joueur dans la liste de dictionnaires
                    clutches_1vx_list.append({"name": clutcher_steamid, f"1v{X}": 1})

# Convertir la liste de dictionnaires en DataFrame pandas
clutches_df = pd.DataFrame(clutches_1vx_list).groupby("name", as_index=False).sum()
print(clutches_df)
```

```python
import os
from demoparser2 import DemoParser
import pandas as pd

# Créer une liste pour stocker les DataFrames de chaque fichier
all_dfs = []

zobrux = ['OzzieOzz','-silentGG','godofbaldz','BELDIYA00','Spiritix']

# Répertoire contenant les fichiers de démonstration
demo_directory = "D:/Python/projet cs v2/demos"

# Parcours des fichiers dans le répertoire
for filename in os.listdir(demo_directory):
    if filename.endswith(".dem"): # Vérifie si le fichier est une démo
        parser = DemoParser(os.path.join(demo_directory, filename))
        max_tick = parser.parse_event("round_end")["tick"].max()
        wanted_fields = ["total_rounds_played","kills_total","assists_total","deaths_total", "mvps", "headshot_kills_total","3k_rounds_total", "4k_rounds_total", "ace_rounds_total" ,"damage_total","utility_damage_total","enemies_flashed_total","alive_time_total"]
        df = parser.parse_ticks(wanted_fields, ticks=[max_tick])

        # Sélection des joueurs spécifiés dans zobrux
        df = df.loc[df['name'].isin(zobrux),:]

        # Ajouter une colonne avec le nom du fichier sans l'extension .dem
        df['Game_id'] = filename[:-4]

        # Ajout du DataFrame à la liste
        all_dfs.append(df)

# Concaténation de tous les DataFrames en un seul
final_df = pd.concat(all_dfs)



final_df
```

```python
rounds_joues = final_df.loc[final_df['name'] == "OzzieOzz", 'total_rounds_played'].sum()

Stats_globales = final_df.rename(columns={"headshot_kills_total":"HS","kills_total":"Kills","assists_total":"Assists","deaths_total":"Deaths","3k_rounds_total":'3K', "4k_rounds_total":'4K', "ace_rounds_total":'5K',"damage_total":"Damages","utility_damage_total":"Utility Damages","enemies_flashed_total":"Flashed Ennemies"})

Stats_globales = Stats_globales.groupby('name').agg({"Kills":"sum","Assists":"sum","Deaths":"sum","HS":'sum',"3K":"sum","4K":"sum","5K":"sum","Damages":"sum","Utility Damages":'sum','Flashed Ennemies':'sum'}).reset_index()

rounds_joues_par_joueur = final_df.groupby('name')['total_rounds_played'].sum()
Stats_globales['Rounds joues'] = Stats_globales['name'].map(rounds_joues_par_joueur)

Stats_globales['HS %'] = round((Stats_globales['HS'] / Stats_globales['Kills']) * 100, 2)
Stats_globales['KPR'] = round((Stats_globales['Kills'] / rounds_joues), 2)
Stats_globales['K/D'] = round((Stats_globales['Kills'] / Stats_globales['Deaths']),2)
Stats_globales['ADR'] = round(Stats_globales['Damages'] / rounds_joues, 2) 

Stats_globales
```
